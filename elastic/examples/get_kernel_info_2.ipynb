```python
from IPython.core.getipython import get_ipython
import os
import json

# connection_file のパスを取得
connection_file = get_ipython().config['IPKernelApp']['connection_file']

# connection_file の内容を表示
if connection_file and os.path.exists(connection_file):
    print(f"Connection File Path: {connection_file}")
    with open(connection_file, 'r') as f:
        data = json.load(f)
        print(json.dumps(data, indent=4))
else:
    print("Connection file not found.")
```
